**#Lowercasing a sentence**

```python
import nltk  #national language toolkit
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
sentence ='It was too far to go to he shop and he did not want her to walk'
sentence = sentence.lower()
sentence_no_stopwards = ' '.join(word for word in sentence.split() if word not in en_stopwords)
print(sentence_no_stopwards)
```

**#Regular Expressions (regex)**

```python
import re
```

```python
my_folder = 'C\Desktop\notes'
print(my_folder)
#Here \n in '\notes' indicates new line

my_folder = r'C\Desktop\notes'
print(my_folder)
#r indicates raw string
```

```python
#re.search() is used to search for a pattern in a given string
#It takes two arguments, the pattern and the string in which the pattern is searched

result_search = re.search('pattern',r'string to search in which contains pattern')
print(result_search)
```

```python
result_search = re.search('pattern is not present',r'string to search in which contains pattern')
print(result_search)
```

```python
#re.sub() is used to replace a pattern
#It takes three arguments, 'old text' , 'new text' , the string

string = r'sara was able to help me find the items I needed quickly'
new_string = re.sub('sara','Sarah',string)
print(new_string)
```

```python
customer_reviews = ['sam was a great help to me in the store',
                    'the cashier was very rude to me, I think her name was eleanor',
                    'amazing work from sadeen',
                    'sarah was able to help me find the items i needed quickly',
                    'lucy is such a great addition to the team',
                    'great service from sara she found me what i wanted']

sarah_reviews = []
pattern_to_find = r'sarah?'

for string in customer_reviews:
  if(re.search(pattern_to_find,string)):
    sarah_reviews.append(string)

print(sarah_reviews)
```

```python
a_reviews = []
pattern_to_find = r'^a' # '^' indicates to search at the start of the string

for string in customer_reviews:
  if(re.search(pattern_to_find,string)):
    a_reviews.append(string)

print(a_reviews)
```

```python
y_reviews = []
pattern_to_find = r'y$' # '$' indicates to search at the end of the string

for string in customer_reviews:
  if(re.search(pattern_to_find,string)):
    y_reviews.append(string)

print(y_reviews)
```

```python
needwant_reviews = []
pattern_to_find = r'(need|want)ed' # '|' indicates to search for either of the two patterns

for string in customer_reviews:
  if(re.search(pattern_to_find,string)):
    needwant_reviews.append(string)

print(needwant_reviews)
```

```python
no_punct_reviews = []
pattern_to_find = r'[^\w\s]'
# '^' stands for not , '\w' stands for word characters , '\s' stands for white spaces

for string in customer_reviews:
  no_punct_string = re.sub(pattern_to_find,'',string)
  no_punct_reviews.append(no_punct_string)

print(no_punct_reviews)
```

**#Stemming**

Reducing words to their base form

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting','connected','connectivity','connect','connects']

for t in connect_tokens:
  print(t , ' : ' ,ps.stem(t))
```

```python
learn_tokens = ['learned','learn','learns','learner','learning','learners']

for t in learn_tokens:
  print(t , ' : ' ,ps.stem(t))
```

```python
likes_tokens = ['likes','worse','better']

for t in likes_tokens:
  print(t , ' : ' ,ps.stem(t))
```

**#N-grams**

To analyze relationship between neighboring words

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = ['the', 'rise', 'of', 'artificial', 'intelligence', 'has', 'led', 'to', 'significant', 'advancements', 'in', 'natural', 'language', 'processing', 'computer', 'vision', 'and', 'other', 'fields', 'machine', 'learning', 'algorithms', 'are', 'becoming', 'more', 'sophisticated', 'enabling', 'computers', 'to', 'perform', 'complex', 'tasks', 'that', 'were', 'once', 'thought', 'to', 'be', 'the', 'exclusive', 'domain', 'of', 'humans', 'with', 'the', 'advent', 'of', 'deep', 'learning', 'neural', 'networks', 'have', 'become', 'even', 'more', 'powerful', 'capable', 'of', 'processing', 'vast', 'amounts', 'of', 'data', 'and', 'learning', 'from', 'it', 'in', 'ways', 'that', 'were', 'not', 'possible', 'before', 'as', 'a', 'result', 'ai', 'is', 'increasingly', 'being', 'used', 'in', 'a', 'wide', 'range', 'of', 'industries', 'from', 'healthcare', 'to', 'finance', 'to', 'transportation', 'and', 'its', 'impact', 'is', 'only', 'set', 'to', 'grow', 'in', 'the', 'years', 'to', 'come']
print(tokens)
```

```python
#unigrams: n=2
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts())
print(unigrams)
```

```python
unigrams[0:10].sort_values().plot.barh(color='lightsalmon', width=.9, figsize=(12, 8))
plt.title('10 Most Frequently Occuring Unigrams')
plt.ylabel('Unigram')
plt.xlabel('# of Occurances')
```

```python
# bigrams: n=2
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print(bigrams)
```

```python

bigrams[0:10].sort_values().plot.barh(color='skyblue', width=.9, figsize=(12, 8))
plt.title('10 Most Frequently Occuring Bigrams')
plt.ylabel('Bigram')
plt.xlabel('# of Occurances')
```

```python
# trigrams: n=3
trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts())
print(trigrams)
```
