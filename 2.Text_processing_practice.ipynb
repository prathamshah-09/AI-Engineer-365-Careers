```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
df = pd.read_csv('tripadvisor_hotel_reviews.csv')
df.head()
```

```python
#lowercase
df['review_lowercase'] = df['Review'].str.lower()
df.head()
```

```python
#stopwords removal
nltk.download('stopwords')
en_stopwords = stopwords.words('english')
en_stopwords.remove('not')
df['review_no_stopwords'] = df['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
df.head()
```

```python
#Punctuation
df['review_no_stopwords_no_punct'] = df \
                                    .apply(lambda x: \
                                    re.sub(r"[*]", \
                                    "star", \
                                    x['review_no_stopwords'] \
                                    ), axis=1)
df.head()
```

```python
#Tokenizing
nltk.download('punkt_tab')
df['tokenized']=df.apply(lambda row: nltk.word_tokenize(row['review_no_stopwords_no_punct']), axis=1)
df.head()
```

```python
#Stemming
ps = PorterStemmer()
df['stemmed'] = df['tokenized'].apply(lambda x: [ps.stem(y) for y in x])
df.head()
```

```python
#Lemmatization
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
df['lemmatized'] = df['tokenized'].apply(lambda x: [lemmatizer.lemmatize(y) for y in x])
df.head()
```

```python
#N-grams
tokens_clean = sum(df['lemmatized'],[])
```

```python
#unigrams
unigrams = (pd.Series(nltk.ngrams(tokens_clean,1)).value_counts())
unigrams.head(10)
```

```python
#bigrams
bigrams = (pd.Series(nltk.ngrams(tokens_clean,2)).value_counts())
bigrams.head(10)
```

```python
ngrams_4 = (pd.Series(nltk.ngrams(tokens_clean,4)).value_counts())
ngrams_4.head(10)
```
