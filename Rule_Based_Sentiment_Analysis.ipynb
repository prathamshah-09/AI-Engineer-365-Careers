```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

TextBlob


```python
from textblob import TextBlob
```

```python
print(sentence_1)
sentiment_score_1 = TextBlob(sentence_1).sentiment.polarity
print(sentiment_score_1)
```

```python
print(sentence_2)
sentiment_score_2 = TextBlob(sentence_2).sentiment.polarity
print(sentiment_score_2)
```

```python
print(sentence_3)
sentiment_score_3 = TextBlob(sentence_3).sentiment.polarity
print(sentiment_score_3)
```

```python
print(sentence_4)
sentiment_score_4 = TextBlob(sentence_4).sentiment.polarity
print(sentiment_score_4)
```
